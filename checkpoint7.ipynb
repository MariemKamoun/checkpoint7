```python
#1
import numpy as np
import pandas as pd
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data=exam_data, index=labels, columns=["name","score","attempts","qualify"])



print('DataFrame initiale est: ')
print("")
print(df)

print(" ")
print("Les trois premières lignes de la DataFrame initiale: ")
print("")
print(df.head(3))

print(" ")
df=df.dropna()
print("DataFrame sans Nan: ")
print("")
print(df)

print(" ")
print("Les colonnes 'name'et 'score': ")
print("")
df_2=df[['name','score']]
print(df_2)

print(" ")
print("DataFrame avec la ligne de Suresh: ")
print("")
df.loc["k"]=["Suresh",15.5,1,"yes"]
print("")
print(df)

print(" ")
print("DataFrame sans la colonne attempts")
print("")
df=df.drop("attempts", axis=1)
print(df)


liste_1=[]
for i in df["score"]:
    if i<10:
        x=0
        liste_1.append(x)
    else:
        x=1
        liste_1.append(x)
print(liste_1)

print("")
print("Dataframe avec colonne success")
print("")
df=df.assign(success=liste_1)
print(df)

print(" ")
print("Dataframe sous forme de csv: ")
print(" ")
df=df.to_csv('mydata')
print(df)
```
